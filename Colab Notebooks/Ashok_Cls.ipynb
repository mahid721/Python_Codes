```python
# Input: nums = [2,7,11,15], target = 9
# Output: [0,1]

# Because nums[0] + nums[1] == 9, we print [0, 1].

# Input: nums = [2,7,11,15], target = 17
# Output: [0,3]
# Because nums[0] + nums[3] == 17, we print [0, 3].



# Given a list of elements, find the indexes of the numbers where num1+num2 == target?
```

```python
# nums = [2,7,11,15]
# target = 9

nums = [2,7,11,15]
target = 22


out_lst = []
for idx in range(len(nums)):
  for idx1 in range(idx+1, len(nums)):
    sum = (nums[idx] + nums[idx1])
    if sum == target:
      out_lst.extend([idx,idx1])
print(out_lst)


```

```python
# Find the maximum of list , minimum of list and Sum of the list  without using Builtin function in python and with using single for loop.


# [2,3,1,4]

# print --> Maximum of the list is 4
# print --> Minimum of the list is 1
# print --> Sum of the list is 10

lst1 = [2,3,1,4]

max = -9999
min = 9999
sum = 0
for i in lst1:
  sum += i
  if i < min:
    min = i
  if i > max:
    max = i
print('min of list : ', min)
print('max of list : ', max)
print('sum of list : ', sum)

```
